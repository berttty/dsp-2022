```python
from simrakit.ProfileParser import Profile
from simrakit.DataProvider import DataProvider

import numpy as np
import matplotlib
from matplotlib import pyplot as plt 
```

```python
dataProvider = DataProvider(local_repo_path="data_cache/")
dataProvider.clone_and_unzip()
```

```python
all_profiles = dataProvider.get_all_profiles()
```

```python
def plot_hist(values, title, xticks, xticklabels, figsize=None):
    if figsize is not None:
        fig, ax = plt.subplots(figsize=figsize)
    else:
        fig, ax = plt.subplots()

    ax.hist(values) 

    if xticks is not None:
        ax.set_xticks(xticks)
        
    if xticklabels is not None:
        ax.set_xticklabels(xticklabels)

    ax.set_title(title) 
```

```python
from simrakit.ProfileParser import BirthYearGroup
# Age Distribution

ages = np.array(list(map(lambda profile: profile.birth_year, all_profiles)))

plot_hist(ages, "Age Distribution", BirthYearGroup.values(), BirthYearGroup.description_titles(), (18,6))
```

```python
from simrakit.ProfileParser import Gender
# Gender Distribution


gender = np.array(list(map(lambda profile: profile.gender, all_profiles)))
plot_hist(gender, "Gender Distribution", Gender.values(), Gender.description_titles())
```

```python
from simrakit.ProfileParser import Experience
# Experience Distribution


experience = np.array(list(map(lambda profile: profile.experience, all_profiles)))
plot_hist(experience, "Experience Distribution in Years", Experience.values(), Experience.description_titles())
```

```python
# Number of Rides distribution


number_of_rides = np.array(list(map(lambda profile: profile.number_of_rides, all_profiles)))
plot_hist(number_of_rides, "Number of Rides per User", None, None)
```
